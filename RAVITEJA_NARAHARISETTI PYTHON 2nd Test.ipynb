1. **(1 pt)** Create a function `vandermonde(n)` which generate the following $n\times n$ array (you can use explicit loops):

$$
V_n = 
\begin{pmatrix}
1^0 & 1^1 & \cdots & 1^n \\
2^0 & 2^1 & \cdots & 2^n \\
\vdots&^  \vdots  & \ddots & \vdots  \\
n^0 & n^1 & \cdots & n^n
\end{pmatrix}
$$
And name `arr3` the array generated by `vandermonde(6)`

**Example:**
```
>>> vandermonde(3)
array([[1, 1, 1],
       [1, 2, 4],
       [1, 3, 9]])
>>> vandermonde(4)
array([[ 1,  1,  1,  1],
       [ 1,  2,  4,  8],
       [ 1,  3,  9, 27],
       [ 1,  4, 16, 64]])
>>> arr3 = vandermonde(6)
```

2. **(1 pt)** Replace all the odd values of `arr3` by -1.

````python
## WRITE IN THIS CELL YOUR CODE WITHOUT ANY ERRORS. YOU SHOULD GENERATE: vandermonde and arr3
## question 1

def vandermode(a):
    va=(np.arange(1,a+1)).reshape(1,a)
    #print(va)
    vb=va**0
    for i in range(1,a):
        vb=np.vstack((vb,va**i))
    
    return vb.T
````

````python
vandermode(6)
arr3=vandermode(6)
print(arr3)
````

````python
## WRITE IN THIS CELL YOUR CODE WITHOUT ANY ERRORS. YOU SHOULD UDPATE THE ARRAY: arr3
## question 2
#arr3[0::2,:]=arr3[0::2,:]-(arr3[0::2,:]+1)
arr3=vandermode(6)
arr3[::2,:]=-1
arr3[:,0]=-1
print(arr3)
#---------------------------------
print("\n")
arr3=vandermode(6)
arr3[arr3%2!=0]=-1
print(arr3)
````
